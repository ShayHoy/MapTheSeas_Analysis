```python
import pandas as pd
import folium
```

```python
Locations = pd.read_csv('/home/mapper/SurveyMonkey/MapTheSeas/AllResponses/Data_All_Responses_180315/CSV/Location_geocode.csv')
Locations = Locations.iloc[:, 1:4]
```

```python
Locations.head()
```

```python
locations = Locations[['Latitude', 'Longitude']]
locationlist = locations.values.tolist()
labels = Locations['Survey Response'].values.tolist()

```

```python
map = folium.Map(location=[38.9, -77.05], tiles='CartoDB positron', zoom_start=11)

marker_cluster = folium.MarkerCluster().add_to(map)

for point in range(0, len(locationlist)):
    folium.Marker(locationlist[point], popup=folium.Popup(labels[point])).add_to(marker_cluster)

map
```

```python
map.save('/home/mapper/SurveyMonkey/MapTheSeas/Locations.html')
```
